```python
from transformers import AutoModelForCausalLM, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("openai-community/gpt2")

# add the EOS token as PAD to aviod warnings
model = AutoModelForCausalLM.from_pretrained("openai-community/gpt2", pad_token_id = tokenizer.eos_token_id)
```

```python
# encode context the generation is conditioned on
model_inputs = tokenizer('I enjoy walking with my dog.', return_tensors='pt')

# generate 40 new tokens
greedy_output = model.generate(**model_inputs, max_new_tokens=40)

print("Output:\n"+100*'-')
print(tokenizer.decode(greedy_output[0], skip_special_tokens=True)) # output should be decode at first to be printed
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

generator = pipeline(task="text-generation", model="openai-community/gpt2")

# Send the text to the pipeline without tokenizing it
generator("Hello, I'm a language model")
```

```python
# Generate single sequence
generator("I enjoy walking with my cute dog.", max_new_tokens=30, num_return_sequences=1)
```

```python
# Generate multiple sequences
generator("I enjoy walking with my cute dog", max_new_tokens=30, num_return_sequences=5)
```
